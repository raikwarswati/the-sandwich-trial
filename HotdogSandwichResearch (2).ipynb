```python
import pandas as pd
import numpy as np

def analyze_dataframe(df):
    analysis = pd.DataFrame({
        'column_name': df.columns,
        'data_type': df.dtypes.values,
        'missing_count': df.isnull().sum().values,
        'missing_percentage': (df.isnull().mean() * 100).round(2).values,
        'unique_values': df.nunique().values
    })

    # Initialize additional columns
    analysis['unique_values_list'] = None
    analysis['min'] = None
    analysis['max'] = None
    analysis['mean'] = None
    analysis['std'] = None

    for index, row in analysis.iterrows():
        col = row['column_name']
        dtype = row['data_type']
        unique_count = row['unique_values']

        # For categorical columns with <10 unique values
        if (dtype == 'object' or unique_count < 10) and unique_count <= 10:
            analysis.at[index, 'unique_values_list'] = df[col].dropna().unique().tolist()

        # For numerical columns
        if np.issubdtype(dtype, np.number):
            analysis.at[index, 'min'] = df[col].min()
            analysis.at[index, 'max'] = df[col].max()
            analysis.at[index, 'mean'] = df[col].mean()
            analysis.at[index, 'std'] = df[col].std()

    return analysis

# Usage
df_analysis = analyze_dataframe(df_text)
print(df_analysis)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def analyze_sentiment_percentages(df, sentiment_col, percentage_col):
    """
    Analyzes sentiment percentage distributions and visualizes them.

    Parameters:
        df (pd.DataFrame): Input dataframe
        sentiment_col (str): Column name containing sentiment categories
        percentage_col (str): Column name containing percentage scores

    Returns:
        pd.DataFrame: Summary statistics
        matplotlib.figure: Boxenplot visualization
    """
    # Generate summary statistics
    sentiment_stats = df.groupby(sentiment_col)[percentage_col].agg(
        ['min', 'max', 'mean', 'median', 'count']
    ).reset_index()

    # Create visualization
    plt.figure(figsize=(12, 6))
    ax = sns.boxenplot(data=df, x=sentiment_col, y=percentage_col, palette='viridis')

    # Add statistical reference lines
    for i, sentiment in enumerate(sentiment_stats[sentiment_col]):
        row = sentiment_stats[sentiment_stats[sentiment_col] == sentiment].iloc[0]
        # ax.hlines(row['min'], i-0.3, i-0.1, color='red', linestyle='--', label='Min' if i == 0 else "")
        # ax.hlines(row['max'], i-0.3, i-0.1, color='blue', linestyle='--', label='Max' if i == 0 else "")
        # ax.hlines(row['mean'], i+0.1, i+0.3, color='green', linestyle='-', label='Mean' if i == 0 else "")
        # ax.hlines(row['median'], i+0.1, i+0.3, color='purple', linestyle='-', label='Median' if i == 0 else "")

    # Customize plot
    plt.title(f"Sentiment Analysis: {percentage_col} Distribution", fontsize=14)
    plt.xlabel("Sentiment Category", fontsize=12)
    plt.ylabel("Percentage Score", fontsize=12)
    plt.legend(title='Statistics')
    plt.grid(axis='y', alpha=0.3)
    sns.despine()

    return sentiment_stats, plt.gcf()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_responses(df, column_name, plot_title):
    """
    Analyzes response distribution for a column and visualizes results.

    Args:
        df (pd.DataFrame): Input dataframe
        column_name (str): Column to analyze
        plot_title (str): Title for plots

    Returns:
        pd.DataFrame: Value counts with percentages
        matplotlib.figure: Bar plot figure
        matplotlib.figure: Donut chart figure
    """


    # Get value counts and percentages
    value_counts = df[column_name].value_counts().reset_index()
    value_counts.columns = ['Response', 'Count']
    percentages = df[column_name].value_counts(normalize=True).mul(100).round(1).reset_index()
    percentages.columns = ['Response', 'Percentage']

    # Create results dataframe
    results = value_counts.merge(
            percentages,
            on='Response'
        )


    # Create bar plot
    plt.figure(figsize=(10, 6))
    bar_ax = sns.barplot(
        data=value_counts,
        x='Response',
        y='Count',
        palette='viridis',
        edgecolor='black'
    )

    # Add annotations
    for p in bar_ax.patches:
        bar_ax.annotate(
            f'{p.get_height():.0f}',
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center',
            va='center',
            xytext=(0, 5),
            textcoords='offset points'
        )

    # Customize bar plot
    bar_ax.set_title(f'Response Counts: {plot_title}', fontsize=16)
    bar_ax.set_xlabel('Response', fontsize=12)
    bar_ax.set_ylabel('Count', fontsize=12)
    sns.despine()
    bar_fig = bar_ax.get_figure()

    # Create donut chart
    plt.figure(figsize=(8, 8))
    wedges, texts, autotexts = plt.pie(
        results['Count'],
        labels=results['Response'],
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette('pastel'),
        wedgeprops={'width': 0.5, 'edgecolor': 'white'},
        textprops={'fontsize': 9}
    )

    # Add center circle
    centre_circle = plt.Circle((0, 0), 0.2, color='white')
    plt.gca().add_artist(centre_circle)

    # Customize donut chart
    plt.title(f'Response Percentages\n{plot_title}', fontsize=14)
    donut_fig = plt.gcf()

    # Close figures to prevent display if not in notebook
    plt.close('all')

    return results, bar_fig, donut_fig

```

```python

# Run analysis
results, bar_plot, donut_chart = analyze_responses(
    df=dft,
    column_name='Q1',
    plot_title='Is a hotdog a sandwich?'
)

# Show results
print("Response Counts and Percentages:")
display(results)

# Display plots
display(bar_plot)
display(donut_chart)
```

## Q1 - Is a hotdog a sandwich vs Generation



```python
# before creating a visualization for GENERATION we need to create a column
# for this using the AGE

# Step 1: Define a function to classify generation from age
def classify_generation(age):
    if age <= 28:
        return 'Gen Z'
    elif age <= 44:
        return 'Millennials'
    elif age <= 60:
        return 'Gen X'
    elif age <= 79:
        return 'Baby Boomers'
    else:
        return 'Silent Generation'

# Step 2: Apply the function to create a new column
df['GENERATION'] = df['AGE'].apply(classify_generation)

```

```python
# Optional: set seaborn theme for prettier plots
sns.set(style="whitegrid", palette="pastel")

# Step 1: Create a crosstab of Q1 responses vs Generation groups
crosstab = pd.crosstab(df['GENERATION'], df['Q1'])

# Step 2: Normalize for proportions (optional)
crosstab_prop = crosstab.div(crosstab.sum(axis=1), axis=0)

# adding colors based on options
# ['❌ No', '✅ Yes', '🤔 It depends', '🚫 I refuse to answer']

q1_colors = {
    '✅ Yes': '#A8E6A3',             # pastel green
    '❌ No': '#FFB3B3',              # pastel red
    '🤔 It depends': '#FFD699',      # pastel orange
    '🚫 I refuse to answer': '#D2B48C'  # pastel brown (tan)
}

# Step 3: Plot count-based grouped bar chart
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='GENERATION', hue='Q1', palette=q1_colors, order=df['GENERATION'].value_counts().index)
plt.title('Is a Hotdog a Sandwich? (by Generation)', fontsize=14, fontweight='bold')
plt.xlabel('Generation')
plt.ylabel('Number of Respondents')
plt.legend(title='Q1 Response')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# Step 4: Display proportion table (optional but useful for storytelling)
print("🔍 Proportion of each Q1 response within each Generation group:\n")
print(crosstab_prop.round(2))
```

1. Older Generations (Boomers & Gen X) are Strongly Against

Their “No” rate is the highest among all groups.

## Q1 vs Q2 (Cereal as Soup?)

```python
# Generate crosstab and visualization for Q1 vs Q2 (Hotdog as Sandwich vs Cereal as Soup)

# Step 1: Crosstab
crosstab_q1_q2 = pd.crosstab(df['Q1'], df['Q2'])
crosstab_q1_q2_prop = crosstab_q1_q2.div(crosstab_q1_q2.sum(axis=1), axis=0).round(2)

# Step 2: Heatmap Visualization
plt.figure(figsize=(8, 5))
sns.heatmap(crosstab_q1_q2_prop, annot=True, cmap="Blues", fmt=".2f", linewidths=0.5)
plt.title("Proportion of Cereal as Soup Opinions (by Hotdog as Sandwich Responses)")
plt.xlabel("Q2: Is Cereal a Soup?")
plt.ylabel("Q1: Is a Hotdog a Sandwich?")
plt.tight_layout()
plt.show()

# Step 3: Styled proportion table
styled_q1_q2_table = crosstab_q1_q2_prop.style\
    .background_gradient(cmap='Blues')\
    .set_caption("Q1 vs Q2: Proportions (Hotdog vs Cereal Debate)")\
    .format("{:.2f}")

styled_q1_q2_table
```

```python
import matplotlib.pyplot as plt

# Use the same proportions from crosstab_q1_q2_prop
df_stack = crosstab_q1_q2_prop.copy()

# --- Stacked Bar Chart ---
plt.figure(figsize=(8, 5))
df_stack.plot(kind='barh', stacked=True, color=['#A8E6A3', '#FFB3B3'], edgecolor='black')

plt.title('How Q1 Responses Align with Q2 (Cereal as Soup)', fontsize=14, fontweight='bold')
plt.xlabel('Proportion')
plt.ylabel('Q1: Is a Hotdog a Sandwich?')
plt.legend(title='Q2: Is Cereal a Soup?', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()


```
